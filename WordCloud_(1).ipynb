```python
#membuka file
import pandas as pd
data = pd.read_excel('top_words_limit11.xlsx')
data.head()
```

```python
df=pd.DataFrame(data[["Word", "P"]])
word = dict(df.values)
word
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(background_color="white", contour_color='steelblue').generate_from_frequencies(word)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
```
